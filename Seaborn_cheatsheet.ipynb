# Seaborn 
The python visulization library is based on matplotlib and provides a high-level interface for drawing attractive statistical graphics.

```python
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

# Distribution plot 
 

```python
sns.distplot(tips['total_bill'],kde= False,bins =20)
```

# Joint plot

```python
sns.jointplot(x='total_bill',y='tip',data = tips,kind='reg')
```

```python
sns.pairplot(tips,hue='sex',palette='spring')
```

```python
sns.rugplot(tips['total_bill'])
```

# Categorical plots

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data= tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data= tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,palette='spring',hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter= True,hue='sex',split=True,palette='spring')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color = 'black')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='violin')
```

# Matrix plots

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
tc =tips.corr()
```

```python
sns.heatmap(tc,annot=True,cmap='spring')
```

```python
fp=flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp,linecolor='white',linewidth=2,cmap='coolwarm')
plt.tight_layout
```

```python
sns.clustermap(fp)
```
