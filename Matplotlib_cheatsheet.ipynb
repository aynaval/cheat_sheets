# Pyplot
Most of the Matplotlib utilities lies under the ***pyoplot*** submodule, and usually imported under the ***plt*** alias

```python
import matplotlib.pyplot as plt
```

```python
xpoints = np.array([1,8])
ypoints = np.array([3,10])

plt.plot(xpoints,ypoints)
```

```python
# plotting without a line
plt.plot(xpoints,ypoints,'o')
```

```python
xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])

plt.plot(xpoints,ypoints)
```

```python
# If x values aren't specified then 0,1,2,3 so on are taken

plt.plot(ypoints)
plt.grid()
```

# Markers 
We can use the keyword ***marker*** to emphasize each point with a specific marker.


```python
plt.plot(ypoints,marker= 'o')
```

```python
plt.plot(ypoints,marker= 'x', markeredgewidth= 18)
```

# Fomrmat String ***fmt***
We can also use shortcut string notation parameter to specify the marker.
This parameter is also called ***fmt***, and is witten this syntax:

Marker | line | color


```python
plt.plot(ypoints,"p:r")
```

```python
plt.plot(ypoints,'s-.k')
```

# Marker size
argument ***markersize*** or ***ms*** is used to set the marker size


```python
plt.plot(ypoints,marker='o',ms =20)
```

```python
plt.plot(ypoints,'s-.k',ms =20)
```

# Marker Color 
Use argument ***markeredgecolor*** or ***mec*** to set color of the color of the edge of the marker.

```python
plt.plot(ypoints,marker= 'o',ms = 20, mec='r')
```

Use keyword argument ***markerfacecolor*** or ***mfc*** to set color inside the edge of the markers

```python
plt.plot(ypoints,marker= 'o',ms = 20, mfc='r')
```

```python
plt.plot(ypoints,marker= 'o',ms = 20, mec='r',mfc='#4CAF50')
```

```python
plt.plot(ypoints,marker= 'o',ms = 20, mec='k',mfc ='hotpink')
```

# linestyle
Use keyword argument ***linestyle*** or ***ls*** to change the style of the plotted line.

```python
plt.plot(ypoints,marker= 'o',ls = 'dashed')
```

```python
plt.plot(ypoints,marker= 'o',ls = '-.')
```

# Line Color 
Use the keyword argument ***color*** or ***c*** to set the color of the line.

```python
plt.plot(ypoints,c='c')
```

```python
plt.plot(ypoints, color = '#4CAF50')
```

# Line Width  
Use the keyword ***linewidth*** or ***lw*** to change the width of the line.
The value is a floating number.

```python
plt.plot(ypoints,lw = 17.2)
```

# Creating Lables for a Plot 
Use ***xlable()*** and ***ylable()*** function.

```python
x = np.arange(10,25,3)
```

```python
x
```

```python
y = x[::-1]
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

# Creating Title for a plot
Use ***title()*** function.

```python
plt.plot(x,y)

plt.title('Sample Plot')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

# Setting font properties for Title and Lables
Use the ***fontdict*** parameter

```python
plt.plot(x,y)

font = {'family':'serif','color':'darkred','size':20 }

plt.title('Sample Plot',fontdict = font)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

# Position the Title
Use ***loc*** parameter in ***title*** to position the title.
Legal values are 'left', 'right' and 'center'.

```python
plt.plot(x,y)

font = {'family':'serif','color':'darkred','size':20 }

plt.title('Sample Plot',fontdict = font, loc = 'left')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

# Adding Grid to a plot 
Use ***grid()*** function 

```python
plt.plot(x,y)
plt.grid()
```

```python
plt.plot(xpoints,ypoints)
plt.grid(axis = 'y')
```

# Setting line properties for the Grid
color = 'color' , linestyle = 'linestyle', linewdth = 'linewidth'

```python
plt.plot(xpoints,ypoints)
plt.grid(axis = 'y',color = 'green', linestyle = '-.', linewidth = 1)
```

# Displaying multiple plots
Use ***subplots()*** function.

```python
# plot 1
plt.subplot(2,2,1)
plt.plot(x,y)

# plot 2
plt.subplot(2,2,2)
plt.plot(xpoints,ypoints)
```

# Super title
The tite for the entire figure, use ***suptitle()*** function

```python
# plot 1
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title('Plot of x and y')

# plot 2
plt.subplot(2,2,3)
plt.plot(xpoints,ypoints)
plt.title('Plot of x-points and y-points')

plt.suptitle("Title of the whole figure")

plt.tight_layout()
```

# Scatter plot

# Colormap
Use keyword ***cmap*** with the value of colormap


```python
colors = np.array([0,100,50,90,80])
plt.scatter(x,y,c =colors)
```

```python
colors = np.array([0,100,50,90,80])
plt.scatter(x,y,c =colors, cmap = 'spring')
plt.colorbar()
```

# Size
To change the size of thr dot use ***s*** argument.

```python
plt.scatter(x,y, s =1000)
```

# Alpha 
To adjust the transparency of the dot use ***alpha*** argument.

```python
plt.scatter(x,y,s =1001,alpha = 0.5)
```

```python
x = np.random.randint(100, size =(80))
y = np.random.randint(100,size=(80))
colors = np.random.randint(500,size=(80))
sizes = 10*np.random.randint(200,size=(80))

plt.scatter(x,y,c = colors,s=sizes,alpha= 0.5,cmap='nipy_spectral')
plt.colorbar()
```
